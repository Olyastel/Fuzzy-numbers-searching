```python
import numpy as np
import pandas as pd
from PIL import Image
from skimage.metrics import structural_similarity as ssim
import os
```

```python
def ssim_upd(path1, path2):
  img1 = np.array(Image.open('/content/drive/MyDrive/Studcamp/for_project/' + path1).resize((32, 18)).convert('1').getdata()).reshape((32, 18))
  img2 = np.array(Image.open(path2).resize((32, 18)).convert('1').getdata()).reshape((32, 18))
  return ssim(img1, img2, data_range=img1.max() - img2.min())
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
image_names = sorted(os.listdir('/content/drive/MyDrive/Studcamp/for_project'))
vssim_upd = np.vectorize(ssim_upd)

similarity_matrix = np.zeros((len(image_names),len(image_names)), dtype=float)

for ind, sym in enumerate(image_names):
  similarity_matrix[ind,:] = vssim_upd(image_names, '/content/drive/MyDrive/Studcamp/for_project/' + sym)
```

```python
index = ['A','B','C','E','H','K','M','O','P','T','Y','X','8','5','4','9','1','7','6','3','2','0']

similarity_matrix_pd = pd.DataFrame(similarity_matrix*100, index=index, columns=index)
similarity_matrix_pd
```

```python
ssim_cut_matrix = np.zeros((len(index),3)).astype(float)
inds_cut_matrix = np.zeros((len(index),3)).astype(str)

for ind, el in enumerate(index):
  ssim_cut_matrix[ind,:] = similarity_matrix_pd.loc[el,:].sort_values(ascending=False).iloc[:3].astype(float)
  inds_cut_matrix[ind,:] = similarity_matrix_pd.loc[el,:].sort_values(ascending=False).iloc[:3].index

ssim_cut_pd = pd.DataFrame(np.concat([ssim_cut_matrix, inds_cut_matrix],axis=1), index=index, columns=['val 1', 'val 2', 'val 3', 'sym 1', 'sym 2', 'sym 3'])
```

```python
ssim_cut_pd = ssim_cut_pd.astype({'val 1':'float64','val 2':'float64','val 3':'float64'})
```

```python
def optimalCSV(theshold, df):
    df = pd.concat([df.iloc[:,:3].where(df.iloc[:,:3] >= theshold, 0),
              df.iloc[:,3:]],axis=1)
    return df
```

```python
ssim_cut_pd = optimalCSV(50, ssim_cut_pd)
ssim_cut_pd
```
